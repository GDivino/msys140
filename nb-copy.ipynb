```python
import pandas as pd

df = pd.read_csv('data.csv')
```

```python
from pandas.api.types import is_numeric_dtype

d = {
    'field': df.columns,
    'count': [df[col].count() for col in df.columns],
    'mean': [df[col].mean() if is_numeric_dtype(df[col]) else None for col in df.columns],
    'median': [df[col].median() if is_numeric_dtype(df[col]) else None for col in df.columns],
    'min': [df[col].min() if is_numeric_dtype(df[col]) else None for col in df.columns],
    'max': [df[col].max() if is_numeric_dtype(df[col]) else None for col in df.columns],
    'std': [df[col].std() if is_numeric_dtype(df[col]) else None for col in df.columns],
    'null_pct' : [df[col].isnull().sum() / len(df[col]) * 100 for col in df.columns]
}

df_profile = pd.DataFrame(d)
df_profile
```

```python
print('Overall record completeness:')
print(round(df['BarangayRes'].notnull().sum() * 100 / len(df['BarangayRes']), 2))
```

```python
date_col = 'DateRepConf'
df[date_col] = pd.to_datetime(df[date_col])
df['month'] = df[date_col].apply(lambda x: x.month)
df['month'].value_counts()
```

```python
month_range = [4, 5, 6, 7, 8, 9]
d = {
    'month': month_range
}

monthly_rc = pd.DataFrame(d)
rc = []
for m in month_range:
    subset = df[df['month'] == m]
    pct = subset['BarangayRes'].notnull().sum() * 100 / len(subset['BarangayRes'])
    rc.append(pct)

monthly_rc['record_completeness'] = rc
monthly_rc
```
